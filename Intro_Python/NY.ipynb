```python
req_df = pd.read_csv("Downloads/311_Service_Requests_from_2010_to_Present.csv.zip", header=0, sep=",", 
            parse_dates=["Created Date", "Closed Date"], dayfirst=True, index_col="Created Date")
```

```python
req_df["Complaint Type"]
```

```python
req_df["Complaint Type"].value_counts()
```

```python
req_df[["Complaint Type", "Borough"]].head()
```

```python
req_df["Complaint Type"].value_counts().plot(kind="bar", figsize=(15,6))
```

```python
req_df["Borough"].value_counts().plot(kind="barh")
```

```python
req_df.plot(kind="scatter", x="Longitude", y="Latitude", figsize=(12,12), alpha=0.1, s=1)
```

```python
req_df[req_df["Borough"] == "MANHATTAN"].plot(kind="scatter", x="Longitude", y="Latitude", figsize=(12,12), alpha=0.1, s=1)
```

```python
colors = ('lime', 'orange', 'm', 'b', 'pink', 'peru')
fig, ax = plt.subplots(1, 1, figsize=(12,12))
for color, barrio in zip(colors, req_df["Borough"].value_counts().index):
    req_df[req_df["Borough"] == barrio].plot(kind="scatter", ax=ax, x="Longitude", y="Latitude",
                                                  figsize=(12,12), alpha=0.1, s=1, c=color)
```

```python
req_df.plot(kind="hexbin", x="Longitude", y="Latitude", gridsize=80, figsize=(14,8), colormap="jet", mincnt=1)
```

```python
req_noise_df = req_df[req_df["Complaint Type"].str.contains("noise", case=False)]
```

```python
req_noise_df.plot(kind="hexbin", x="Longitude", y="Latitude", gridsize=80, 
                  figsize=(14,8), colormap="jet", mincnt=1)
```

```python
req_df[req_df["Complaint Type"] == "Drinking"].plot(kind="hexbin", x="Longitude", y="Latitude", gridsize=80, 
                  figsize=(14,8), colormap="jet", mincnt=1)
```

```python
req_df[req_df["Descriptor"] == "Loud Talking"].plot(kind="hexbin", x="Longitude", y="Latitude", gridsize=80, 
                  figsize=(14,8), colormap="jet", mincnt=1)
```
